```python
import pandas as pd
```

```python
#membaca dataset dan mengubahnya menjadi dataframe
data = pd.read_csv('Social_Network_Ads.csv')
```

```python
#mengubah dataset menjadi dataframe pandas
df = pd.read_csv('Social_Network_Ads.csv')
```

```python
#untuk melihat 5 baris pertama dataset
df.head()
```

```python
#untuk melihat nila kosong pada setiap atribut
df.info()
```

```python
#drop pada kolom yang tidak diperlukan (id dihilangkan))
data = df.drop(columns=['User ID'])
data
```

```python
#menjalankan proses one-hot encoding dengan pd.get_dummies()
data = pd.get_dummies(data)
data
```

```python
#pisahkan atribut dan label
predictions = ['Age', 'EstimatedSalary', 'Purchased', 'Gender_Female', 'Gender_Male']
X=data[predictions]
y=data['Purchased']
```

```python
#melakukan normalisasi terhadap data
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X)
scaled_data = scaler.transform(X)
scaled_data = pd.DataFrame(scaled_data, columns=X.columns)
scaled_data.head()
```

```python
from sklearn.model_selection import train_test_split

#bagi data menjadi train dan test untuk setiap atribut dan label
X_train, X_test, y_train, y_test = train_test_split(scaled_data, y, test_size=0.2, random_state=1)
```

```python
from sklearn import linear_model

#latih model dengan fungsi fit
model = linear_model.LogisticRegression()
model.fit(X_train, y_train)
```

```python
from sklearn.preprocessing import StandardScaler

#uji akurasi model
model.score(X_test, y_test)
```
